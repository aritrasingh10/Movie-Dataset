```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
```

## Importing CSV file

```python
df = pd.read_csv('movies.csv')
df.head()
```

## Finding missing data

```python
df.isnull().sum()
```

## Deleting all the missing value rows

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df.head(3)
```

## Creating a column which has the correct year of release

```python
df['year_rl'] = df['released'].str.split(', ',expand=True)[1].str.split(' ',expand=True)[0]
```

```python
df.tail()
```

## Top 5 highest gross films

```python
df.sort_values(by=['gross'],ascending=False).head()
```

```python
df.sort_values(by=['score'],ascending=False).head()
```

## Top 5 highest Budget films
